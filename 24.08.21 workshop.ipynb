```python
menu = {
    "espresso": {
        "ingredients": {
            "water": 50,
            "coffee": 18,
        },
        "cost": 1.5,
    },
    "latte": {
        "ingredients": {
            "water": 200,
            "milk": 150,
            "coffee": 24,
        },
        "cost": 2.5,
    },
    "cappuccino": {
        "ingredients": {
            "water": 250,
            "milk": 100,
            "coffee": 24,
        },
        "cost": 3.0,
    }
}
```

```python
def money(req):
    while True:
        money = float(input('Your money: $'))
        if enough_money(req, money):
            sources['money']+=menu[req]['cost']
            break
```

```python
def enough_money(chosen, money):
    if menu[chosen]['cost'] > money:
        print("Sorry that's not enough money. Money refunded.")
        return False
    elif menu[chosen]['cost'] < money:
        print(f"Here is ${money-menu[chosen]['cost']} dollars in change.")
        return True
    else: return True
```

```python
sources = {
    "water": 1000,
    "milk": 1000,
    "coffee": 1000,
    "money": 100
}
```

```python
while True:
    while True:
        req = input('What would you like? (espresso/latte/cappuccino): ').strip().lower()
        if req in ('off', 'report', 'espresso', 'latte', 'cappuccino'):
            break
        else: print('Your choice is invalid, please try again!')
            
    if req == 'off':
        print('Sorry, the machine is currently under maintenance')
        break
    if req == 'report':
        print(f"water: {sources['water']}ml", f"milk: {sources['milk']}ml", f"coffee: {sources['coffee']}gr", f"money: ${sources['money']}", sep = '\n')
        break
    
    
    if req == 'espresso':
        if sources['water'] > 50 and sources['coffee'] > 18:
            sources['water']-=50
            sources['coffee']-=18
            money(req)
            print('Here is your espresso. Enjoy!')
        else:
            if not sources['water'] > 50:
                print('Sorry there is not enough water')
            if not sources['coffee'] > 18:
                print('Sorry there is not enough coffee') 
            break
    elif req == 'latte':
        if sources['water'] > 200 and sources['coffee'] > 24 and sources['milk'] > 150:
            sources['water']-=200
            sources['coffee']-=24
            sources['milk']-=150
            money(req)
            print('Here is your latte. Enjoy!')
        else:
            if not sources['water'] > 200:
                print('Sorry there is not enough water')
            if not sources['coffee'] > 24:
                print('Sorry there is not enough coffee') 
            if not sources['milk'] > 150:
                print('Sorry there is not enough milk')
            break
    elif req == 'cappuccino':
        if sources['water'] > 250 and sources['coffee'] > 24 and sources['milk'] > 100:
            sources['water']-=250
            sources['coffee']-=24
            sources['milk']-=100
            money(req)
            print('Here is your cappuccino. Enjoy!')
        else:
            if not sources['water'] > 250:
                print('Sorry there is not enough water')
            if not sources['coffee'] > 24:
                print('Sorry there is not enough coffee') 
            if not sources['milk'] > 100:
                print('Sorry there is not enough milk')
            break
    
```
