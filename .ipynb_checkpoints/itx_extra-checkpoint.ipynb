**QUEST II -** Write a Python program to find index position and value of the maximum and minimum values in a given list of numbers using lambda.

```python
ls = [6, 7, 9, 2, 4, 3]
min_index = lambda x : min(x, key = x.index)
max_index = lambda x : max(x, key = x.index)
print(min_index(ls), ls[min_index(ls)])
print(max_index(ls), ls[max_index(ls)])
```

```python
def equal(*arg):
    return lambda x : max(list(arg), key = list(arg).index)
res = equal(1,2,3,2)
res()
```

**QUEST III -** Write a function that inserts a white space between every instance of a lower character followed immediately by an upper character.

**Expected Output:**

func1("WhiteSpacesBetweenLowerandUppercaseLetters") âžž "Spaces Between Lower and Uppercase Letters"

```python
def func1(s):
    s1 = s[0]
    for i in range(1, len(s)):
        if s[i].isupper():
            s1+=' '+s[i]
        else:
            s1+=s[i]
    return s1
            
func1("WhiteSpacesBetweenLowerandUppercaseLetters")        
```
