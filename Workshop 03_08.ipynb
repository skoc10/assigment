 1: Random Lottery Pick. Generate 100 random lottery tickets and pick two lucky tickets from it as a winner.
Note you must adhere to the following conditions:

The lottery number must be 10 digits long.
All 100 ticket number must be unique.

Hint: Generate a random list of 1000 numbers using randrange() and then use the sample() method to pick lucky 2 tickets.



sample output:"Lucky 2 lottery tickets are [6764548492, 3615771612]"

```python
import random
ls = []
while len(ls) != 100:
    ls.append(random.randrange(1000000000, 10000000000))
    ls = list(set(ls))
print(random.sample(ls, 2))
```

3. Write a Python program to get next day of a given date.
(don't forget leap year)

Expected Output:

Input a year: 2016                                                      
Input a month [1-12]: 08                                                
Input a day [1-31]: 23                                                  
The next date is [yyyy-mm-dd] 2016-8-24  

```python
year = int(input('year: '))
month = int(input('month: '))
day = int(input('day: '))

if (year % 400 == 0):
    leap_year = True
elif (year % 100 == 0):
    leap_year = False
elif (year % 4 == 0):
    leap_year = True
else:
    leap_year = False

if month in (1, 3, 5, 7, 8, 10, 12):
    month_length = 31
elif month == 2:
    if leap_year:
        month_length = 29
    else:
        month_length = 28
else:
    month_length = 30
    
if day < month_length:
    day += 1
else:
    day = 1
    if month == 12:
        month = 1
        year += 1
    else:
        month += 1
print(f'The next date is [yyyy-mm-dd] {year}-{month}-{day}')
```

4. Write a function that takes an ordered list of numbers (a list where the elements are in order from smallest to largest) and another number. The function decides whether or not the given number is inside the list and returns (then prints) an appropriate boolean.

Extras:Use binary search.

output : "True" or "False" 

```python
def inside(ls, n):
    if n in ls:
        return True
    else:
        return False
```

```python
inside([1,2,3,5,6,9,11,55,88], 9)
```

5. Write a program determine if a sentence is a pangram.

What is Pangram? ... The sentence formed by using all the letters of the alphabet is called a "pangram".

The quick brown fox jumps over a lazy dog.

```python
sentence = input('Give me a sentence: ').strip(' ,.!?').lower()
set1 = {'q','w','e','r','t','y','u','i','o','p','a','s','d','f','g','h','j','k','l','z','x','c','v','b','n','m'}
set2 = set(sentence)
set2.remove(' ')
set2.remove(',')
if set2 == set1:
    print('Your sentence is a Pangram')
else:
    print('Your sentence is not a Pangram')
```
